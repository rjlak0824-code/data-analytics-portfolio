# FoodHub Order Demand Analysis

An exploratory data analysis of food delivery orders across New York City restaurants, uncovering demand patterns, customer behavior, and operational insights to support business growth.

## Business Context

FoodHub is a food aggregator platform that connects customers in New York with their favorite restaurants through a single app. With the city's restaurant count growing rapidly and busy professionals increasingly relying on food delivery, understanding order patterns is critical to improving customer experience and operational efficiency.

This analysis explores FoodHub's order dataset to identify key trends in restaurant demand, cuisine preferences, delivery performance, and customer satisfaction.

## Setup: Import Libraries

```python
# import libraries for data manipulation
import numpy as np
import pandas as pd

# import libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Data Overview

We begin by loading the dataset and understanding its structure, data types, and quality.

The dataset contains **1,898 orders** across 9 columns. Each row represents a single food delivery order placed through the FoodHub platform.

The dataset has **no null values**, meaning all records are complete and ready for analysis.

### Statistical Summary

We examine the distribution of key numeric variables — food preparation time, delivery time, and order cost.

```python
df[['cost_of_the_order', 'food_preparation_time', 'delivery_time']].describe().round(2)
```

### 2.4 High-Performing Restaurants

Identifying restaurants with both high order volume and strong ratings to inform promotional partnerships.

```python
# Restaurants with rating count > 50 and average rating > 4
df_rated = df[df["rating"] != "Not given"].copy()
df_rated["rating"] = pd.to_numeric(df_rated["rating"])
top_rated = df_rated.groupby("restaurant_name").filter(lambda x: len(x) > 50 and x["rating"].mean() > 4)
print("High-volume, high-rated restaurants:")
print(top_rated["restaurant_name"].unique())
```

**Blue Ribbon Fried Chicken** and **Blue Ribbon Sushi** stand out as both high-volume and high-rated — strong candidates for featured placement or promotional partnerships.

### 2.5 Revenue & Commission Analysis

FoodHub charges 25% commission on orders above $20, and 15% on orders at or below $20.

```python
commission_high = df[df["cost_of_the_order"] > 20]["cost_of_the_order"].sum() * 0.25
commission_low = df[df["cost_of_the_order"] <= 20]["cost_of_the_order"].sum() * 0.15
total_revenue = commission_high + commission_low
print(f"Revenue from orders > $20 : ${commission_high:,.2f}")
print(f"Revenue from orders <= $20: ${commission_low:,.2f}")
print(f"Total platform revenue    : ${total_revenue:,.2f}")
```

Total estimated platform revenue across all orders is **$6,166.30**. High-value orders (>$20) contribute disproportionately — optimizing average order value would directly increase revenue.

#### Orders Taking More Than 60 Minutes End-to-End

```python
df["total_time"] = df["food_preparation_time"] + df["delivery_time"]
pct_over_60 = (df["total_time"] > 60).mean() * 100
print(f"Orders taking more than 60 minutes: {pct_over_60:.2f}%")
```

**10.54% of orders** take over 60 minutes from placement to delivery. This minority but significant segment likely drives negative ratings and churn — a key area for operational improvement.

#### Weekday vs Weekend Delivery Time Comparison

```python
weekday_avg = df[df["day_of_the_week"] == "Weekday"]["delivery_time"].mean()
weekend_avg = df[df["day_of_the_week"] == "Weekend"]["delivery_time"].mean()
print(f"Mean delivery time - Weekdays: {weekday_avg:.2f} mins")
print(f"Mean delivery time - Weekends: {weekend_avg:.2f} mins")
print(f"Difference: {abs(weekday_avg - weekend_avg):.2f} mins")
```

Weekday deliveries average **28.34 minutes** vs **22.47 minutes** on weekends — nearly 6 minutes faster. Weekday demand may be straining delivery capacity, pointing to a need for better driver allocation during business hours.

## 3. Key Business Insights

## Key Business Insights

- **Weekend demand is higher** than weekdays, with American cuisine being the most ordered — targeted weekend campaigns could amplify this further.
- **Top 5 restaurants** (Shake Shack, The Meatball Shop, Blue Ribbon Sushi, etc.) account for a disproportionate share of orders — platform health depends on retaining these partners.
- **38.8% of orders are unrated**, making it difficult to measure true customer satisfaction — improving the post-delivery rating prompt is a quick win.
- **10.54% of orders exceed 60 minutes** total delivery time — this segment is a key driver of poor customer experience and likely correlates with low ratings.
- **Weekday deliveries are ~6 minutes slower** than weekends, suggesting capacity constraints during peak work-hour ordering.
- **29.24% of orders exceed $20**, generating higher commissions — strategies to increase average order value (bundles, upsells) would directly improve revenue.

## Business Recommendations

**1. Optimize Weekday Delivery Capacity**
Allocate more delivery drivers during weekday lunch and evening peaks to close the 6-minute gap with weekend delivery times.

**2. Promote High-Performing Restaurant Partners**
Feature Blue Ribbon Fried Chicken, Blue Ribbon Sushi, and Shake Shack prominently in the app — their combination of volume and ratings makes them trust signals for new customers.

**3. Improve Post-Order Rating Collection**
With 38.8% of orders unrated, a simple push notification or in-app nudge post-delivery could significantly improve feedback data quality and coverage.

**4. Address Long-Delivery Orders Proactively**
For the 10.54% of orders exceeding 60 minutes, implement real-time alerts to customers and consider partial compensation (discount vouchers) to protect satisfaction scores.

**5. Weekend Cuisine Promotions**
Since American, Japanese, and Italian cuisines dominate weekends, run targeted cuisine-specific promotions on Fridays to build weekend order momentum.

**6. Loyalty Program for High-Frequency Customers**
Customers like 52832 and 47440 are placing 10+ orders. A tiered loyalty program would increase retention among this high-value segment and encourage similar behavior in mid-frequency users.
